# Distribución nomal mediante método Box-Muller

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import random 
import pandas as pd
sns.set()
```

### Se generan valores aleatorios x,y con distribucion normal estandar mediante metodo de Box-Muller

```python
u1 = []
u2 = []
for i in range(100000): #Se gneran 100,000 valores aleatorios 
    a = random.uniform(0, 1)                  
    b = random.uniform(0, 1)
    u1.append(a)
    u2.append(b)
u1=np.array(u1)
u2=np.array(u2)
        
X = np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2) #valores aleatorios x
Y = np.sqrt(-2*np.log(u1))*np.sin(2*np.pi*u2) #valores aleatorios y
```

### Se genera tabla de los valores aleatorios

```python
datos = {"u1":u1, "u2":u2 , "x":X, "y":Y}
valores = pd.DataFrame(datos)
print("Primeros 5 valores aleatorios")
print(valores.head(5)) #Imprimimos solo los primeros 5
print(" ")
```

### Descripcion de los valores aleatorios obtenidos por metodo Box-Muller

```python
print("Descripcion de valores simulados mediante Box-Muller")
print(valores.describe())
```

### Hacemos una lista de datos de las coordenadas de la caminata alaetoria

```python
xi, yi =0, 0 #Iniciamos desde el origen la caminata
camino_X=[]
camino_Y=[]
for i in range(len(X)):
    xi=xi+X[i] 
    yi=yi+Y[i]    
    camino_X.append(xi)
    camino_Y.append(yi)
```

### Grafica de movimiento aleatorio en plano xy

```python
plt.plot(camino_X,camino_Y, "-", markersize = 0.0001)
plt.ylabel('y')
plt.xlabel('x')
plt.title('Partícula Browniana')
plt.show()
```
